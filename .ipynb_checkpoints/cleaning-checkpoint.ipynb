```python
import re

# Supprimer les emojis d'un texte.
def remove_emojis(text):
    if isinstance(text, str):
        emoji_pattern = re.compile("["
                                   u"\U0001F600-\U0001F64F"  # emoticons
                                   u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                                   u"\U0001F680-\U0001F6FF"  # transport & map symbols
                                   u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                                   u"\U00002702-\U000027B0"  # miscellaneous
                                   u"\U000024C2-\U0001F251" 
                                   "]+", flags=re.UNICODE)
        return emoji_pattern.sub(r'', text)
    else:
        return text

# Supprimer les emojis de la colonne 'Description'
df['Description'] = df['Description'].apply(remove_emojis)
df
```

```python
# Remplacer les valeurs manquantes de la colonne 'Description' par la chaÃ®ne 'has no description'
df['Description'].fillna('has no description', inplace=True)
```

```python
# Convertir la colonne 'Date creation' en objets datetime en utilisant le format spÃ©cifiÃ©
df['Date creation'] = pd.to_datetime(df['Date creation'], format='%Y-%m-%dT%H:%M:%SZ')

# Obtenir les annÃ©es distinctes
distinct_years = df['Date creation'].dt.year.unique()
distinct_years = sorted(distinct_years)

# Afficher les annÃ©es distinctes
print(distinct_years)
```
